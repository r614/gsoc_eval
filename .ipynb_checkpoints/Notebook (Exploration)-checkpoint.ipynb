```python
from datetime import datetime
from datetime import timezone
import pytz
import os

fileName = os.listdir('/home/roshan/Projects/CERN/gsoc_eval/')[4]
timeString = fileName[:18]
timeInt = int(timeString)
timeInt = timeInt/(pow(10,9))

utcTimeString = datetime.utcfromtimestamp(timeInt).strftime('%Y-%m-%dT%H:%M:%SZ')
utcTime = datetime.utcfromtimestamp(timeInt)
cernZone = pytz.timezone('Europe/Zurich')
cernTime = utcTime.astimezone(cernZone)

print(utcTime)
print(cernTime)
```

```python
import h5py

fileInput = h5py.File(fileName)
print(list(fileInput['AwakeEventData'].keys()))

def print_attrs(name, obj):
    print(name)
    for key, val in obj.attrs.iteritems():
        print("    %s: %s" % (key, val))


fileInput.visititems(print_attrs)
```

```python
%matplotlib inline
import numpy as np
from scipy.signal import medfilt
import matplotlib.pyplot as plt


path = fileInput['AwakeEventData']['XMPP-STREAK']['StreakImage']
imageData = path['streakImageData']
imageHeight = path['streakImageHeight']
imageWidth = path['streakImageWidth']



print(imageData)
print(imageHeight)
print(imageWidth)

imageArray = np.reshape(imageData, (imageWidth[0], imageHeight[0]))
img = medfilt(imageArray)
imgplot = plt.imshow(img)
```
